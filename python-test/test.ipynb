```python
# Read file
def read_file(file_path):
    with open(file_path, 'r') as f:
        lines = (f.readlines())
    return lines
```

```python
# Store all files
files = []
# Store file titles 
file_titles = []
# Populate both arrays
for file in os.listdir():
    file_path = f"{path}/{file}"
    file_titles.append(f"{file}")
    files.append(read_file(file_path))

# Print one file
# for one_file in files:
#     print(one_file)

# Print file titles array
print(file_titles)
```

```python
# word stripping code
def stripper(word):
    chars = []
    final_word = ""
    for x in range(len(word)):
        chars.append(word[x])
    for c in chars:
        if c.isalnum():
            final_word += c
    return final_word
```

```python
# Read through files array and create dictionary
words_dict = {}

class InstancesInFile:
    def __init__(self, file_idx, line_num):
        self.file_idx = file_idx
        # self.line_nums = set()
        # self.line_nums.add(line_num)
        self.line_num = line_num
    
    # def add_instance(self, line_num):
    #     self.line_nums.add(line_num)

    def print_instance(self):
        result = "File Index: " + str(self.file_idx) + "; Line Number: " + str(self.line_num)
        return result

def process_line(line, file_idx, line_idx):
    # print("processing line: " + line)
    for word in line.split():
        word = word.lower()
        # need to strip words here
        word = stripper(word)
        if word in words_dict.keys():
            # print("word found")
            new_instance = InstancesInFile(file_idx, line_idx)
            words_dict[word].append(new_instance)
        else:
            new_instance = InstancesInFile(file_idx, line_idx)
            new_list = [new_instance]
            # new_list.append(new_instance)
            words_dict.update({word: new_list})
        # print(words_dict[word])

for file_idx, file in enumerate(files):
    for line_idx, line in enumerate(file):
        process_line(line, file_idx, line_idx)
```

```python

# Print values in dictionary
for value in words_dict.values():
    print("value: ")
    print(value)
    for instance in value:
        print("instance: ")
        print(instance.print_instance())
```
